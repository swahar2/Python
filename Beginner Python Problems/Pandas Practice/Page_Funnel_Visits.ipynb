Import all the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])

checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

Step 1: Inspect the DataFrames using `print` and `head`

```python
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))
```

Step 2: Left merging visits and cart

```python
visits_cart = pd.merge(visits, cart, how = 'left')
print(visits_cart.head())
```

Step 3: How long is `visits_cart`?

```python
visits_cart_length = len(visits_cart)

print(visits_cart_length)
print(visits_cart['user_id'].count())
```

Step 4: How many timestamps are null for `cart_time`?

```python
cart_null = visits_cart['cart_time'].isnull().sum()
print(cart_null)
```

Step 5: What percentage only visited?

```python
percentage_only_visited = (cart_null / visits_cart_length) * 100
print(percentage_only_visited)
```

Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
cart_checkout = pd.merge(cart, checkout, how = 'left')
print(cart_checkout.head())
```

```python
cart_checkout_length = len(cart_checkout)
print(cart_checkout_length)
```

```python
checkout_null = cart_checkout['checkout_time'].isnull().sum()
print(checkout_null)
```

```python
percentage_no_checkout = (checkout_null / cart_checkout_length) * 100
print(percentage_no_checkout)
```

Step 7: Merge it all together

```python
all_data = visits.merge(cart, how = 'left').merge(checkout, how = 'left').merge(purchase, how = 'left')
print(all_data.head())
```

Step 8: % of users who got to checkout but did not purchase

```python
checkout_purchase = pd.merge(checkout, purchase, how = 'left')
print(checkout_purchase.head())
```

```python
checkout_length_df = all_data[~all_data['checkout_time'].isnull()]
checkout_length = len(checkout_length_df)
print(checkout_length)
```

```python
purchase_null_df = all_data[all_data['purchase_time'].isnull() & ~all_data['checkout_time'].isnull()]
purchase_null = len(purchase_null_df)
print(purchase_null)
```

```python
percentage_no_purchase = (purchase_null / checkout_length) * 100
print(percentage_no_purchase)
```

Step 9: check each part of the funnel, let's print all 3 of them again

```python
print(percentage_only_visited)
print(percentage_no_checkout)
print(percentage_no_purchase)
```

*The weakest part of the funnel is clearly getting a person who visited the site to add a tshirt to their cart. Once they've added a t-shirt to their cart it is fairly likely they end up purchasing it. A suggestion could be to make the add-to-cart button more prominent on the front page.*


Step 10: adding new column

```python
all_data['time_to_purchase'] = all_data['purchase_time'] - all_data['visit_time']
print(all_data.head())
```

Step 11: examine the results

```python
print(all_data['time_to_purchase'])
```

Step 12: average time to purchase

```python
print(all_data['time_to_purchase'].mean())
```
